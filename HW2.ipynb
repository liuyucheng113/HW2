### **Arrays**

- A numpy array is a grid of values, all of the same type, and is indexed by a tuple of nonnegative integers.

- The number of dimensions is the rank of the array;

- The shape of an array is a tuple of integers giving the size of the array along each dimension.

### **Array Operations**

```python
# create 2 2d-array
import numpy as np

arr_2d_1 = np.array([[1, 2, 3], [4, 5, 6]])
arr_2d_2 = np.array([[7, 8, 9], [10, 11, 12]])
print("First 2D array:")
print(arr_2d_1)

print("\nSecond 2D array:")
print(arr_2d_2)

# Elementwise sum; both produce an array
arr_sum=np.add(arr_2d_1,arr_2d_2)

print("\nElementwise sum of the two 2D arrays:")
print(arr_sum)
```

```python
# Elementwise difference; both produce an array
import numpy as np

arr_2d_1 = np.array([[1, 2, 3], [4, 5, 6]])
arr_2d_2 = np.array([[7, 8, 9], [10, 11, 12]])

print("First 2D array:")
print(arr_2d_1)

print("\nSecond 2D array:")
print(arr_2d_2)

arr_diff=np.subtract(arr_2d_1,arr_2d_2)

print("\nElement-wise difference of the two 2D arrays:")

print(arr_diff)
```

```python
# Elementwise division; both produce an array
import numpy as np

arr_2d_1 = np.array([[10, 20, 30], [40, 50, 60]])
arr_2d_2 = np.array([[2, 4, 6], [8, 10, 12]])


arr_div = np.divide(arr_2d_1, arr_2d_2)

print("First 2D array:")
print(arr_2d_1)

print("\nSecond 2D array:")
print(arr_2d_2)

print("\nElement-wise division of the two 2D arrays:")
print(arr_div)
```

```python
# Elementwise product; both produce the array

import numpy as np

arr_2d_1 = np.array([[1, 2, 3], [4, 5, 6]])
arr_2d_2 = np.array([[7, 8, 9], [10, 11, 12]])

arr_pro=np.multiply(arr_2d_1,arr_2d_2)

print("First 2D array:")
print(arr_2d_1)

print("\nSecond 2D array:")
print(arr_2d_2)

print("\nElement-wise product of the two 2D arrays:")
print(arr_pro)
```

```python
# Dot product or matrix multiplication
arr_2d_1 = np.array([[1, 2, 3], [4, 5, 6]])

# Create the second 2D array (matrix)
arr_2d_2 = np.array([[7, 8],  [9, 10],  [11, 12]])

print("First 2D array (matrix):")
print(arr_2d_1)

print("Second 2D array (matrix):")
print(arr_2d_2)

arr_dot=np.dot(arr_2d_1,arr_2d_2)
print("\nDot product")
print(arr_dot)
```
